```python
times_df = pd.read_csv("timesData.csv")
times_df
```

```python
from fractions import Fraction
import numpy as np
```

```python
times_df = times_df.replace('-', 0)
times_df
```

```python
int_students = times_df['international_students'].str.strip('%').astype(float)/100
int_students
```

```python
times_df['int_students'] = int_students
times_df
```

```python
times_df = times_df.drop(['international_students'], axis = 1)
times_df
```

```python
times_df[['female','male']] = times_df["female_male_ratio"].str.split(" : ",expand=True,)
times_df
```

```python
times_df = times_df.drop(['female_male_ratio'], axis = 1)
times_df
```

```python
times_df = times_df.fillna(0)
times_df
```

```python
times_df['female'] = pd.to_numeric(times_df['female'])
```

```python
times_df['male'] = pd.to_numeric(times_df['male'])
```

```python
times_df.dtypes
```

```python
times_df['female_male'] = np.where(times_df['male'] < 1, times_df['male'], times_df['female']/times_df['male'])
times_df
```

```python
times_df = times_df.drop(['female'], axis = 1)
times_df.head()
```

```python
times_df = times_df.drop(['male'], axis = 1)
times_df.head()
```

```python
df5 = times_df.loc[(times_df['year'] >= 2012) & (times_df['year'] <= 2015)]
df5
```

```python
df5.to_csv('2012_2015_times.csv', index = False)
```

```python
import sqlite3
  
conn = sqlite3.connect(r'C:\Users\Thomas\Documents\dataproject\times.db')

times_data = pd.read_csv('2012_2015_times.csv')
# Write the data to a sqlite table
times_data.to_sql('times_rankings', conn, if_exists='replace', index=False)
  
# Create a cursor object
cur = conn.cursor()
# Fetch and display result
for row in cur.execute('SELECT * FROM times_rankings'):
    print(row)
# Close connection to SQLite database
conn.close()
```
